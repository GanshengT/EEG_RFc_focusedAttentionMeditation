```python
#This function is aimed to train a general RFc, and retrieve the performance
#and report important features

#parameters definition
subject_list=c("BOOU","CACA","DEER","EZLE","GRCL","LENO","LOVA","MATA","MOHU","NIJO","PAMA","PRBA","TRWI","ZAMI")
#setwd("E:/important file/CENTRALE/BCI ET EEG/df_withPAC_matrix")
df_path = "D:/20182019EEG_CS/df_withPAC_matrix"
# setwd()

# fc
multMerge = function(mypath){
  filenames = list.files(path = mypath, full.names = TRUE)
  datalist = lapply(filenames, 
                    function(x){read.csv(file = x,
                                         header = TRUE,
                                         stringsAsFactors = FALSE)})
  Reduce(function(x,y) {merge(x, y, all = TRUE)}, datalist)
}

df = multMerge(df_path)
df$MediOrStory<-factor(ifelse(df$MediOrStory%in%c('0'),'0','1')) 
df$time=NULL
df$HR=NULL
df$SC=NULL
df$ST=NULL
df$Resp=NULL
df$HRV=NULL
for(i in 1:21){
    df[paste0(sub("D.*","",colnames(df)[4*i]),"TBR")]=df[4*i-1]/df[4*i-2]
    df[paste0(sub("D.*","",colnames(df)[4*i]),"ATR")]=df[4*i-3]/df[4*i-1]
}

write.csv(df,"D:/20182019EEG_CS/df_withPAC_matrix/df_global.csv")
```

```python
df <- read.table("D:/20182019EEG_CS/df_withPAC_matrix/df_global.csv",header = TRUE,sep=",")[, -1]
df$MediOrStory<-factor(ifelse(df$MediOrStory%in%c('0'),'0','1')) 
library(randomForest)
Model_rf =randomForest:: randomForest(MediOrStory~., data=df)
FeatImpt=randomForest:: importance(Model_rf)
FeatImpt=FeatImpt/sum(FeatImpt[,1])
FeatImpt
```

```python
df <- read.table("D:/20182019EEG_CS/df_withPAC_matrix/df_global.csv",header = TRUE,sep=",")
train_idx = sample(1:nrow(df),0.8*nrow(df))
test_idx = setdiff(1:nrow(df),train_idx)
df_train = df[train_idx,]
df_test = df[test_idx,]
library(randomForest)
Model_rf =randomForest:: randomForest(MediOrStory~., data=df_train)
FeatImpt=randomForest:: importance(Model_rf)
num_feature=5
FeatImptSorted=sort(FeatImpt)
FeatImptOrdered=rev(order(FeatImpt))
FeatImptOrdered=FeatImptOrdered[1:num_feature]
FeatureUsed=c()
FeatureUsed_impt=c()
FeatureUsed_rank=c()
j=1
for (i in FeatImptOrdered){
FeatureUsed=c(FeatureUsed,row.names(FeatImpt)[i])
FeatureUsed_impt=c(FeatureUsed_impt,FeatImpt[i])
FeatureUsed_rank=c(FeatureUsed_rank,j)
j=j+1
}
rm(j)
Feature_cloud=cbind(FeatureUsed,FeatureUsed_impt,FeatureUsed_rank)
write.csv(Feature_cloud,"D:/20182019EEG_CS/global_RFc_results/feature_cloud.csv")
```

```python
# performance
library(caret)
rfFrml <- as.formula(paste("MediOrStory", paste(row.names(FeatImpt)[FeatImptOrdered],sepA = "", 
                                              collapse = " + "), sep = " ~ "))

tc <- trainControl(method = "cv", 10)
modFitP <- train(rfFrml, method = "rf", data = df_train, trControl = tc, allowParallel=TRUE, importance=TRUE, ntree = 500 )
print(modFitP)
prdval=predict(modFitP, df_test)
groudTrue_test = df_test$MediOrStory
cfsMatrix=confusionMatrix(groudTrue_test, prdval,positive='1')
```
